```python
# frame extraction function
def getFrame(cam,sec,count,char):
    cam.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    hasFrames,image = cam.read()
    w_path = 'ds/'+str(char)
    if hasFrames:
        cv2.imwrite(os.path.join(w_path,char+str(count)+'.jpg'),image)
    return hasFrames
```

```python
# capture image A

char = 'A'

cam_a_1 = cv2.VideoCapture('vid/A_1.mp4')
sec = 0
frameRate = 0.1 # capture image in each 0.1 seconds
count = 1
success = getFrame(cam_a_1,sec,count,char)
while success:
    if count<250:
        count += 1
        sec += frameRate
        sec = round(sec,2)
        success = getFrame(cam_a_1,sec,count,char)
    else:
        break

cam_a_2 = cv2.VideoCapture('vid/A_2.mp4')
sec = 0
frameRate = 0.1 # capture image in each 0.1 seconds
count = 251
success = getFrame(cam_a_2,sec,count,char)
while success:
    if count<500:
        count += 1
        sec += frameRate
        sec = round(sec,2)
        success = getFrame(cam_a_2,sec,count,char)
    else:
        break
```
